```python
data_dict = []

for row, j in stealing_df.iterrows():
    catcher_id = j['credit_id']
    
    runner_id = j['pitcher_id']
    
    catcher_name = j['credit']
    
    runner_name = j['runner_name']
    
    date = j['Date']
    
    try:
    
        runner_df = pd.read_csv(f'Runner_final/{runner_name}.csv')

        catcher_df = pd.read_csv(f'catchers/{catcher_name}.csv')
        
        
    except:
        print(catcher_name)
        continue
        
    if 'Exchange' not in catcher_df.columns:
        continue
        
    catcher_df.drop(columns=['Unnamed: 0'], inplace=True)

    catcher_df.drop_duplicates(inplace=True)

    catcher_past = catcher_df.loc[catcher_df['Date'] < date].sort_values(by='Date').tail(10)

    if len(catcher_past) < 10:
        continue

    catcher_past['Result'] = catcher_past['Action'].apply(lambda x : 1 if x =='Stolen' else 0)
    
    uniques = catcher_past['Arm_Strength'].unique()




    if catcher_past['Arm_Strength'].dtypes == 'object':
    
        arm = catcher_past['Arm_Strength'].tolist()
        numbers = [item.strip() for item in arm if item != ' --']

        if len(numbers) == 0:
            catcher_past['Arm_Strength'].replace(' --', np.nan, inplace=True)
            catcher_past['Arm_Strength'] = catcher_past['Arm_Strength'].astype(float)
            
        else:
            
            repl = float(numbers[0])
            catcher_past['Arm_Strength'].replace(' --', repl, inplace=True)
            catcher_past['Arm_Strength'] = catcher_past['Arm_Strength'].astype(float)

        
    if catcher_past['Exchange'].dtypes == 'object':
    
        arm = catcher_past['Exchange'].tolist()
        numbers = [item.strip() for item in arm if item != ' --']
        if len(numbers) == 0:
            catcher_past['Exchange'].replace(' --', np.nan, inplace=True)
            catcher_past['Exchange'] = catcher_past['Exchange'].astype(float)
            
        else:
            
            repl = float(numbers[0])
            catcher_past['Exchange'].replace(' --', repl, inplace=True)
            catcher_past['Exchange'] = catcher_past['Exchange'].astype(float)

    sprints = []

    for row, i in catcher_past.iterrows():

        runner = i['runner_name']

        current_year = i['Year'] 
        
        if len(runners_final_df.loc[(runners_final_df['runner_name'] == runner) & (runners_final_df['Current_Year'] == current_year)]['Sprint']) == 0:
            sprints.append(np.nan)
            continue

        sprint = runners_final_df.loc[(runners_final_df['runner_name'] == runner) & (runners_final_df['Current_Year'] == current_year)]['Sprint'].tolist()[0]

        sprints.append(sprint)
    catcher_past['Sprint'] = sprints

    mean_catch = catcher_past[['Result', 'Exchange' ,'Arm_Strength', 'Sprint']].mean()

    Result_catcher = mean_catch['Result']

    Exchange_catcher = mean_catch['Exchange']

    Arm_Strength_catcher = mean_catch['Arm_Strength']

    Sprint_catcher = mean_catch['Sprint']
    
    runner_df.drop(columns=['Unnamed: 0'], inplace=True)

    runner_df.drop_duplicates(inplace=True)

    runner_past = runner_df.loc[runner_df['Date'] < date].sort_values(by='Date').tail(10)
    
    if len(runner_past) < 10:
        continue

    runner_past['Result'] = runner_past['Action'].apply(lambda x : 1 if x =='Stolen' else 0)

    Exchanges=[]
    Arm_Strengths = []

    for row, i in runner_past.iterrows():

        catcher = i['credit']

        current_year = i['Year']
        
        if len(catchers_final_df.loc[(catchers_final_df['credit'] == catcher) & (catchers_final_df['Current_Year'] == current_year)]) == 0:
            Exchanges.append(np.nan)
            Arm_Strengths.append(np.nan)
            continue

        Exchange = catchers_final_df.loc[(catchers_final_df['credit'] == catcher) & (catchers_final_df['Current_Year'] == current_year)]['Exchange'].tolist()[0]

        Arm_Strength = catchers_final_df.loc[(catchers_final_df['credit'] == catcher) & (catchers_final_df['Current_Year'] == current_year)]['Arm_Strength'].tolist()[0]

        Exchanges.append(Exchange)
        
        Arm_Strengths.append(Arm_Strength)

    runner_past['Exchange'] = Exchanges

    runner_past['Arm_Strength'] = Arm_Strengths
    
    if runner_past['Arm_Strength'].dtypes == 'object':
    
        numbers = [item.strip() for item in Arm_Strengths if item != ' --']
        repl = float(numbers[0])
        runner_past['Arm_Strength'].replace(' --', repl, inplace=True)

        runner_past['Arm_Strength'] = runner_past['Arm_Strength'].astype(float)

    mean_catch = runner_past[['Result', 'Exchange' ,'Arm_Strength', 'Sprint']].mean()

    Result_runner = mean_catch['Result']

    Exchange_runner = mean_catch['Exchange']

    Arm_Strength_runner = mean_catch['Arm_Strength']

    Sprint_runner = mean_catch['Sprint']
    
    data_dict.append({
        'Sprint_runner': Sprint_runner,
        'Exchange_runner': Exchange_runner,
        'Result_runner': Result_runner,
        'Arm_Strength_runner': Arm_Strength_runner,
        'Result_catcher': Result_catcher, 
        'Exchange_catcher': Exchange_catcher,
        'Arm_Strength_catcher': Arm_Strength_catcher,
        'Sprint_catcher': Sprint_catcher,
        'Date': date,
        'Runner_id': runner_id,
        'Result': j['Result']
        
        
        
    })
```

```python
stolen_df = pd.DataFrame(data_dict)
stolen_df.to_csv('stolen_df_mean_10.csv')
```
