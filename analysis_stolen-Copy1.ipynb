```python
from itertools import combinations, product
lr = [0.01, 0.02, 0.005, 0.05]

mf = ["sqrt", "log2", 0.5, 0.8]

md = [5, 10, 15, 20, 30, 50, 80]

n_e = [150, 200, 300, 400, 500, 1000]

min_sp = [2, 3, 5, 10]



combos = list(product(lr, mf, md, n_e ,min_sp))

len(combos)
```

```python
results = {}

# Splitting into Train and Test sets.
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=78)

# Creating a StandardScaler instance.
scaler = StandardScaler()
# Fitting the Standard Scaler with the training data.
X_scaler = scaler.fit(X_train)

from sklearn.ensemble import GradientBoostingClassifier

# Scaling the data.
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)

for combo in combos:


    classifier = GradientBoostingClassifier(n_estimators=combo[3],
    learning_rate=combo[0],
    max_features=combo[1],
    max_depth=combo[2],
    min_samples_split=combo[4],
    random_state=0)
    classifier.fit(X_train_scaled, y_train.ravel())
    
    # Make predictions on test data
    y_pred = classifier.predict(X_test_scaled)
    
    precision = precision_score(y_test, classifier.predict(X_test_scaled))
    
    recall = recall_score(y_test, y_pred)

    
    results[combo] = {'Training': f"{classifier.score(X_train_scaled, y_train)}",'Precision':precision, 'Recall': recall}
```

```python
results_df = pd.DataFrame.from_dict(results, orient='index').sort_values(by='Precision', ascending=False)
results_df.head(10)
```

```python
results_df['Max Depth'] = results_df.apply(lambda row: row.name[2], axis =1)
results_df['Min Feat'] = results_df.apply(lambda row: row.name[1], axis =1)
results_df['N'] = results_df.apply(lambda row: row.name[-1], axis =1)
```

```python
results_df.loc[(results_df['Recall'] >= 0.2) & (results_df['Max Depth'] <= 50)].head(10)
```
