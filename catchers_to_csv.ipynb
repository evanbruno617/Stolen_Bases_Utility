```python
folder_path = 'stolen_bases_new'
```

```python
csv_files = [file for file in os.listdir(folder_path) if file.endswith('.csv')]
```

```python
csv_files
```

```python
csv_list = []

for file in csv_files:
    file_path = os.path.join(folder_path, file)
    df = pd.read_csv(file_path)
    csv_list.append(df)

mlb_df = pd.concat(csv_list, ignore_index=True)
mlb_df.to_csv('combined.csv', index=False)
```

```python
cleaned_mlb = mlb_df[[col for col in mlb_df.columns if col not in ['Unnamed: 0.1', 'Unnamed: 0',
                                                    'Respon_Pitcher']]]
```

```python
catcher_names = cleaned_mlb['credit'].unique().tolist()

for catcher in catcher_names:
    catcher_df = cleaned_mlb.loc[cleaned_mlb['credit']==catcher]
    clean_catcher_name = str(catcher).replace(" ", "_") # can't have spaces between the names yo
    catcher_df.to_csv(f"{clean_catcher_name}.csv", index=False)
```
